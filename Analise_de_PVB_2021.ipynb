# Análise dos ultimos seis meses.

```python
dados2 = dados.loc[:,['jun_20','jul_20','agos_20','set_20','out_20','nov_20','dez_20','jan_21', 'fev_21','març_21','abr_21']]
dados2
```

```python
dados.index
```

```python
itens_A = dados.loc[['PVB INCOLOR 0.38MM 2.4','PVB INCOLOR 0.38MM 2.2'],['jun_20','jul_20','agos_20','set_20','out_20','nov_20','dez_20','jan_21', 'fev_21','març_21','abr_21']].T
```

```python
itens_A.plot(kind = 'line', figsize = (12,8), title = 'Movimentação de PVB', xlabel = 'Mês', ylabel = 'Rolos')
```

```python
itens_A.plot(kind = 'bar', figsize = (12,8), title = 'Movimentação de PVB', xlabel = 'Mês', ylabel = 'Rolos')
```

```python
pvb1 = sns.boxplot(data= [itens_A['PVB INCOLOR 0.38MM 2.4']], orient = 'h', palette='hls')
pvb1.set_title('PVB INC 0.38MM 2.44', fontsize = 10)
pvb1.set_xlabel('Rolos', fontsize = 10)
pvb1
```

```python
pvb2 = sns.boxplot(data= [itens_A['PVB INCOLOR 0.38MM 2.2']], orient = 'h', palette='hls')
pvb2.set_title('PVB INC 0.38MM 2.22', fontsize = 10)
pvb2.set_xlabel('Rolos', fontsize = 10)
pvb2.set_ylabel('PVB', fontsize = 10)
pvb2
```

```python
total = dados2.sum(axis =1)
total
```

```python
media_mes = dados2.mean(axis = 1)
media_mes
```

```python
desviopad_mes = dados2.std(axis = 1)
desviopad_mes
```

```python
mediana_mes =dados2.median(axis = 1)
mediana_mes
```

```python
q1 = dados2.quantile(0.25, axis = 1)
q1
```

```python
q3 = dados2.quantile(0.75, axis = 1)
q3
```

```python

mov_seis_meses = pd.DataFrame({"Total Rolos":total, "Media Mês":media_mes,"Primeiro Quartil": q1,"Mediana":mediana_mes,"Terceiro Quartil": q3,"Desvio padrão": desviopad_mes}) 
mov_seis_meses
```
