```python
import pandas as pd
```

###**Preparando dados.**


```python
url= "https://pt.wikipedia.org/wiki/Lista_de_campe%C3%B5es_da_Ta%C3%A7a_dos_Campe%C3%B5es_Europeus_e_Liga_dos_Campe%C3%B5es_da_UEFA"
```

```python
html = pd.read_html(url, match="País")
html
```

```python
html_campeos = html[0]
html_campeos
```

```python
html_campeos.tail()
```

```python
tab_campeao = html_campeos[['Campeão','Edição']].loc[:67]
tab_campeao
```

```python
tab_campeao['Valor'] = 1
tab_campeao
```

```python
tabela_campeao = tab_campeao.groupby(['Campeão','Edição']).sum()['Valor'].unstack().T.fillna(0)
tabela_campeao
```

```python
tabela_campeao.index
```
