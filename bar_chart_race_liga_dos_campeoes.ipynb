```python
#Importando os pacotes
import pandas as pd
```

###**Preparando dados.**


```python
#dados para o projeto
url= "https://pt.wikipedia.org/wiki/Lista_de_campe%C3%B5es_da_Ta%C3%A7a_dos_Campe%C3%B5es_Europeus_e_Liga_dos_Campe%C3%B5es_da_UEFA"
```

```python
html = pd.read_html(url, match="País")
html
```

```python
#selecionando a tabela referente ao titulos dos times
html_campeos = html[0]
html_campeos
```

```python
html_campeos.tail()
```

```python
#separando as informações para construção da tabela
tab_campeao = html_campeos[['Campeão','Edição']].loc[:69]
tab_campeao
```

```python
#Substituir a coluna "Campeão" removendo os parênteses e tudo dentro deles
tab_campeao['Campeão'] = tab_campeao['Campeão'].str.replace(r'\s*\(.*?\)', '', regex=True)
tab_campeao
```

```python
#atribuindo valor ao titulos
tab_campeao['Valor'] = 1
tab_campeao
```

```python
#construindo as tabela final
tabela_campeao = tab_campeao.groupby(['Campeão','Edição']).sum()['Valor'].unstack().T.fillna(0)
tabela_campeao
```
